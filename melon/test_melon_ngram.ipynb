````python
# n그램 생성 함수 정의 

def generate_ngrams(text, n):
    tokens = text.split()  # 단어 단위로 분리
    ngrams = [tuple(tokens[i:i + n]) for i in range(len(tokens) - n + 1)]
    return ngrams
````

````python
# 1. 별도 모듈 없이 n그램 생성 함수 활용 했을 때 

## case 1. n=2
ngram_test_1 = generate_ngrams(song1, 2)
ngram_test_1
````

````python
import nltk

# n-gram 계산 
unigrams = nltk.ngrams(song1,1)
bigrams  = nltk.ngrams(song1,2)
trigrams  = nltk.ngrams(song1,3)

# 결과 출력
print("Unigrams:", unigrams)
print("Bigrams:", bigrams)
print("Trigrams:", trigrams)
````

````python
# 단어 예측 

from nltk import ngrams
from collections import defaultdict

# 학습 데이터
text = song1

# n-그램 계산
tokens = text.split()
bigrams = list(ngrams(tokens, 2))

# n-그램 모델 생성
model = defaultdict(list)

for w1, w2 in bigrams:
    model[w1].append(w2)

# "나는" 이후 예측 단어
next_word_candidates = model["나는"]

# 예측된 단어 출력
if next_word_candidates:
    next_word = next_word_candidates[0]  # 첫 번째 후보 단어 선택
    print("다음 단어:", next_word)
else:
    print("다음 단어를 예측할 수 없습니다.")
````

````python
# 문장 유사도 테스트 

song2 = df2['lylics'][15] # 두번째 노래가 같아서 (중복) 그 다음 노래로 할게요 

def jaccard_similarity(sentence1, sentence2):
  n_grams1 = set(ngrams(sentence1.split(), 2))
  n_grams2 = set(ngrams(sentence2.split(), 2))
  intersection = n_grams1 & n_grams2
  union = n_grams1 | n_grams2
  return len(intersection) / len(union)

# 문장 유사도 계산
similarity = jaccard_similarity(song1, song2)

# 결과 출력
print("문장 유사도:", similarity)

### 이렇게 하는건 의미가 없는것 같습니다 
````

````python
#텍스트 분류 

from nltk import ngrams
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# 학습 데이터
texts = [
  ("난 왜", "의문"),
  ("그 미움까지 그리운 걸까", "의문"),
  ("그토록 미웠던 끝 모습을 꺼내보고 말야", "후회"),
  ("그때 그렇게 있지 말고 조금 더 기대볼걸", "후회"),
  ("그대만 있으면", "기대"),
  ("시간이 지나면 우릴 담은 기억 모두 다 흐릿해지겠죠", "기대"),
  ("조금만 더 기억할게요", "결심"),
  ("모른 채로 듣지 않고 뒤돌지 않은 체 어떻게든 그렇게 지나가볼게요", "결심")
]

# N-gram 기반 특징 추출
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
features = vectorizer.fit_transform([text for text, _ in texts])

# 모델 학습
labels = [label for _, label in texts]
model = LogisticRegression().fit(features, labels)

# 새로운 텍스트 분류
new_text = "그때도 여전히 나는"
new_features = vectorizer.transform([new_text])
prediction = model.predict(new_features)

# 결과 출력
print("분류 결과:", prediction[0])
````

### bert 테스트 

````python
# BERT https://blog.hashscraper.com/bert-text-sentiment-analysis-tutorial/

import torch
from transformers import BertTokenizer, BertForSequenceClassification
import torch.nn.functional as F
````

````python
model_name = "kykim/bert-kor-base"
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=2)
tokenizer = BertTokenizer.from_pretrained(model_name)
````

````python
def classify_emotion(text):
    # 텍스트 토큰화 및 패딩
    tokens = tokenizer(text, padding=True, truncation=True, return_tensors="pt")

    # 예측 수행
    with torch.no_grad():
        prediction = model(**tokens)

		# 예측 결과를 바탕으로 감정 출력
    prediction = F.softmax(prediction.logits, dim=1)
    output = prediction.argmax(dim=1).item()
    labels = ["부정적", "긍정적"]
    print(f'[{labels[output]}]\\n')
````

````python
def predict_sentence(song):
    classify_emotion(song)
````
